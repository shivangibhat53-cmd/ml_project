### **<u>3.8 Adding columns for "Total Score" and "Average"</u>**

```python
df['total_score'] = df['math_score']+df['reading_score'] + df['writing_score']
df['average'] = df['total_score']/3
```

```python
df.head(3)
```

```python
#Find the number of students that scored full marks in each subjects
reading_full = df[df['reading_score']==100]['average'].count()
writing_full = df[df['writing_score']==100]['average'].count()
maths_full   = df[df['math_score']==100]['average'].count()

print('Number of students with full marks in reading - ',reading_full)
print('Number of students with full marks in writing - ',writing_full)
print('Number of students with full marks in math    - ',maths_full)
```

```python
# Count of students with marks less than 20
reading_lt_20 = df[df['reading_score']<20]['average'].count()
writing_lt_20 = df[df['writing_score']<20]['average'].count()
maths_lt_20   = df[df['math_score']<20]['average'].count()

print('Number of students with marks less than 20 in reading - ',reading_lt_20)
print('Number of students with marks less than 20 in writing - ',writing_lt_20)
print('Number of students with marks less than 20 in math    - ',maths_lt_20)
```

#### **INSIGHTS**
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section
